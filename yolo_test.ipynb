## Yolo Pose

```python
from yolo import *
import cv2

# Detect and Seg
DS_NMS_THRESH = 0.45
DS_OBJ_THRESH = 0.25

#Seg
MAX_DETECT = 300

#Pose and OBB
PO_NMS_THRESH = 0.4
PO_OBJ_THRESH = 0.5


#Common
DATASET = COCO
IMG_SIZE = 320

KEY_ESCAPE = 27

if __name__ == '__main__':
    RK3588_RKNN_MODEL = "./pose/yolo11n-pose-RK3588_320x320_i8.rknn"
    IMG_PATH = './pose.jpg'
    '''
    # Yolo(
    # TASK, 
    # RK3588_RKNN_MODEL, 
    # input_size = 640, 
    # DATASET = COCO, 
    # DS_NMS_THRESH = 0.45, 
    # DS_OBJ_THRESH = 0.25, 
    # PO_NMS_THRESH = 0.4, 
    # PO_OBJ_THRESH = 0.5, 
    # MAX_DETECT = 300
    #)
    '''
    yolo = Yolo( "pose", RK3588_RKNN_MODEL, input_size = IMG_SIZE, DATASET=DATASET, CALC_ANGLE=True)
    img_org = cv2.imread(IMG_PATH)
    predbox = yolo.pose(img_org)
    img = yolo.pose.draw(img_org, predbox, box_vis = False, angle=False)
    cv2.imshow("post process result", img)
    cv2.waitKey(0)
    # Release
    cv2.destroyAllWindows()
    yolo.release()
```

## Yolo Seg

```python
from yolo import *
import cv2

# Detect and Seg
DS_NMS_THRESH = 0.45
DS_OBJ_THRESH = 0.25

#Seg
MAX_DETECT = 300

#Pose and OBB
PO_NMS_THRESH = 0.4
PO_OBJ_THRESH = 0.5


#Common
DATASET = GARBAGE
IMG_SIZE = 640

KEY_ESCAPE = 27

if __name__ == '__main__':
    if DATASET == CRACK:
        RK3588_RKNN_MODEL = './seg/yolov11_crack/yolo11n-crack-RK3588_i8.rknn'
        IMG_PATH = './crack.jpg'
    if DATASET == GARBAGE:
        RK3588_RKNN_MODEL = './seg/yolov11_garbage/yolov11n_taco-RK3588_i8.rknn'
        IMG_PATH = './garbage.jpg'
    '''
    # Yolo(
    # TASK, 
    # RK3588_RKNN_MODEL, 
    # input_size = 640, 
    # DATASET = COCO, 
    # DS_NMS_THRESH = 0.45, 
    # DS_OBJ_THRESH = 0.25, 
    # PO_NMS_THRESH = 0.4, 
    # PO_OBJ_THRESH = 0.5, 
    # MAX_DETECT = 300
    #)
    '''
    yolo = Yolo( "seg", RK3588_RKNN_MODEL, input_size = IMG_SIZE, DATASET=DATASET)
    img_org = cv2.imread(IMG_PATH)
    boxes, classes, scores, seg_img = yolo.seg(img_org)
    if boxes is not None:
        img_p = yolo.seg.merge_seg(img_org, seg_img, classes)
        yolo.seg.draw(img_p, boxes, scores, classes, "all", label=True)
    else:
        print( "Box is None.")

    cv2.imshow("post process result", img_p)
    cv2.waitKey(0)

    #show_rgb_image(img_org, 'post process result')

    cv2.destroyAllWindows()
    yolo.release()
```
